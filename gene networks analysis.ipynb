```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the gene network
gene_net = pd.read_csv("example_data/gene networks/chrom.txt", sep=" ", encoding='utf-8', header=None)
gene_net.columns = ["gene1", "gene2", "weight"]
```

```python
# Count the number of associations for each gene
gene_degree = gene_net.groupby('gene1').size().reset_index().rename(columns={0: 'degree'})
gene_degree = gene_degree.sort_values(by='degree', ascending=True)
gene_degree_distribution = gene_degree['degree'].value_counts().reset_index()
gene_degree_distribution.columns = ["Count", "Degree"]
gene_degree_distribution
```

```python
plt.clf()
# è®¾ç½®å­—ä½“ä¸ºä»¿å®‹
plt.rcParams['font.family'] = 'Arial'
# ç»Ÿä¸€è®¾ç½®è½´åˆ»åº¦çš„å­—ä½“å¤§å°
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
# ç»Ÿä¸€è®¾ç½®æ ‡é¢˜ã€xyè½´åç§°çš„å­—ä½“å¤§å°
plt.rcParams['axes.titlesize'] = 25
plt.rcParams['axes.labelsize'] = 25
# è®¾ç½®å›¾ä¾‹çš„å­—ä½“å¤§å°
plt.rcParams['legend.fontsize'] = 22
plt.rcParams['axes.unicode_minus'] = False  # æ˜¾ç¤ºæ­£ã€è´Ÿçš„é—®é¢˜

plt.figure(figsize=(6, 5), dpi=600)
ax = plt.axes()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
sns.histplot(gene_degree_distribution, x="Degree", color="#a17db4", bins=100)

# Decoration
plt.xlabel("Number of gene")
plt.ylabel("Count")
# plt.savefig("example_data/gene networks/chrom degree.jpg", dpi=600, bbox_inches='tight')
plt.show()
```
